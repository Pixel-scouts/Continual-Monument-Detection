```python
old={'Chyasimdeval':346,
        'HariShankhar Temple':355,
        'Garud Statue':372,
        'Bhimsen Temple':58,
        'Mani Ganesh Temple':171,
        'Mani Mandap':249,
        'Krishna Temple PDS':522,
        'Taleju Bell pds':247,
        'Taleju temple north':280,
        'Taleju temple south':209,
        'Vishwanath Temple':598,
        'Yog Narendra Malla Statue': 113,
        'Char Narayan Temple':328,
       }
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = list(old.keys())
data = {
    'old': list(old.values()),
    'new': [item1 + item2 for item1, item2 in zip(list(new.values()), list(old.values()))]
}
print(labels)
print(data)

```

```python
x = np.arange(len(labels))  # the label locations
width = 0.4  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(figsize=(12, 6),layout='constrained')

for sets, values in data.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, values, width, label=sets)
    ax.bar_label(rects, padding=5)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('number of instances', fontweight='bold')
ax.set_xlabel('monuments', fontweight='bold')
ax.set_title('old and new data')
ax.set_xticks(x+width/2 , labels)

plt.xticks(rotation=90) 
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 900)
plt.savefig("data_addition.png", dpi=None, bbox_inches='tight', format='png')
plt.show()
```
