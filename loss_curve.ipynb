```python
import pandas as pd
file_path=r"D:\Major\meta learning\metaaaaaaaaa\train_loss_8 (2).csv"
df = pd.read_csv(file_path)
# len(df.iloc[0,:])
first_row = list(df.head(1))
# print(first_row)
print(len(first_row))
print(type(first_row[0]))
# df.head(0)
```

```python
itr_per_epoch=(len(first_row)//8)
print(itr_per_epoch)
```

```python
len(first_row)
losses=[]
for num in first_row:
    try:
        losses.append(float(num))
    except:
        print(num)
print(len(losses))
```

```python
import numpy as np
```

```python
avg_loss_per_epoch=[]
for i in range(0,len(losses),itr_per_epoch):
    loss_in_epoch=losses[i:i+itr_per_epoch]
#     print(loss_in_epoch)
    summ=sum((loss_in_epoch))
    avg=summ/itr_per_epoch
    avg_loss_per_epoch.append(avg)
if(len(avg_loss_per_epoch))>8:
    l1=avg_loss_per_epoch.pop()
    l2=avg_loss_per_epoch.pop()
    avg_loss_per_epoch.append(l1+l2)
```

```python
print(avg_loss_per_epoch)
```

```python
import matplotlib.pyplot as plt
epochs = list(range(1, len(avg_loss_per_epoch) +1))
plt.plot(epochs, avg_loss_per_epoch, marker='o', linestyle='-', color='b')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Train Loss vs Epoch')
plt.xticks(epochs)
plt.grid(True)
plt.savefig("train_loss.png", dpi=None, bbox_inches='tight', format='png')
plt.show()
```
